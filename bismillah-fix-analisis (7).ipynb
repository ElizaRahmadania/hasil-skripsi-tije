```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

## Load Data 10 Menit Tap-in

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (15, 6)

# LOAD DATA
file_path = '/kaggle/input/datasets/elizarahmadania/tapin-tj-40-hari/(1) 40 HARI TAP IN 10 MENIT.xlsx'
df_raw_10 = pd.read_excel(file_path)

# TRANSFORMASI KE LONG FORMAT
kolom_waktu = df_raw_10.columns[2:]
df_long_10 = pd.melt(
    df_raw_10,
    id_vars=['Kode Halte', 'Nama Halte'],
    value_vars=kolom_waktu,
    var_name='Interval_Waktu',
    value_name='Jumlah_Penumpang'
)

# PEMBERSIHAN NAMA KOLOM (Menghapus .1, .2, ... .39)
df_long_10['Interval_Waktu'] = df_long_10['Interval_Waktu'].str.replace(r'\.\d+$', '', regex=True)

# AGREGASI (JUMLAHKAN)
df_agregat_10 = df_long_10.groupby(['Kode Halte', 'Nama Halte', 'Interval_Waktu'])['Jumlah_Penumpang'].sum().reset_index()

# TAHAP PEMBERSIHAN DESIMAL (UBAH KE INTEGER)

df_agregat_10['Jumlah_Penumpang'] = pd.to_numeric(df_agregat_10['Jumlah_Penumpang'], errors='coerce').fillna(0).astype(int)

# TAMPILKAN DATA AGREGASI (HEAD) DALAM FORMAT KE SAMPING
df_final_wide_10 = df_agregat_10.pivot_table(
    index=['Kode Halte', 'Nama Halte'],
    columns='Interval_Waktu',
    values='Jumlah_Penumpang',
    aggfunc='sum' # Menggunakan sum untuk memastikan data tetap angka bulat
).reset_index()

# Menghilangkan nama index pada kolom agar tampilan bersih
df_final_wide_10.columns.name = None

# Tambahan: Pastikan semua kolom jam juga bertipe integer (biar gak ada .0 di tampilan)
kolom_jam = df_final_wide_10.columns[2:]
df_final_wide_10[kolom_jam] = df_final_wide_10[kolom_jam].fillna(0).astype(int)

print("Tampilan 5 baris pertama data agregasi selama 40 hari:")
display(df_final_wide_10.head())
```

```python
import pandas as pd
import numpy as np

# Data mapping halte ke koridor
data_mapping = [
    ('Blok M', 'Koridor 1'), ('CSW', 'Koridor 1'), ('ASEAN', 'Koridor 1'),
    ('Kejaksaan Agung', 'Koridor 1'), ('Masjid Agung', 'Koridor 1'),
    ('Bundaran Senayan', 'Koridor 1'), ('Senayan Bank Jakarta', 'Koridor 1'),
    ('Polda Metro Jaya', 'Koridor 1'), ('Bendungan Hilir', 'Koridor 1'),
    ('Karet', 'Koridor 1'), ('Dukuh Atas', 'Koridor 1'), ('Tosari', 'Koridor 1'),
    ('Bundaran HI Astra', 'Koridor 1'), ('MH Thamrin', 'Koridor 1'),
    ('Kebon Sirih Arah Blok M', 'Koridor 1'), ('Kebon Sirih Arah Kota', 'Koridor 1'),
    ('Monumen Nasional', 'Koridor 1'), ('Harmoni Arah Blok M', 'Koridor 1'),
    ('Sawah Besar Arah Blok M', 'Koridor 1'), ('Taman Sari', 'Koridor 1'),
    ('Glodok', 'Koridor 1'), ('Kota', 'Koridor 1'), ('Kali Besar', 'Koridor 1'),
    ('Museum Sejarah Jakarta', 'Koridor 1'),
    ('Pulo Gadung', 'Koridor 2'), ('Bermis', 'Koridor 2'), ('Pulo Mas', 'Koridor 2'),
    ('Perintis Kemerdekaan', 'Koridor 2'), ('Pedongkelan', 'Koridor 2'),
    ('Cempaka Mas', 'Koridor 2'), ('Sumur Batu', 'Koridor 2'),
    ('Cempaka Baru', 'Koridor 2'), ('Pasar Cempaka Putih', 'Koridor 2'),
    ('Rawa Selatan', 'Koridor 2'), ('Galur', 'Koridor 2'), ('Senen Raya', 'Koridor 2'),
    ('RSPAD', 'Koridor 2'), ('Pejambon', 'Koridor 2'), ('Gambir', 'Koridor 2'),
    ('Istiqlal', 'Koridor 2'), ('Juanda', 'Koridor 2'), ('Pecenongan', 'Koridor 2'),
    ('Monumen Nasional', 'Koridor 2'), ('Balai Kota', 'Koridor 2'), ('Gambir 2', 'Koridor 2'),
    ('Kwitang', 'Koridor 2'),
    ('Monumen Nasional', 'Koridor 3'), ('Petojo', 'Koridor 3'), ('Roxy', 'Koridor 3'),
    ('Grogol', 'Koridor 3'), ('Jelambar', 'Koridor 3'), ('Damai', 'Koridor 3'),
    ('Taman Kota', 'Koridor 3'), ('Jembatan Gantung', 'Koridor 3'),
    ('Pulo Nangka', 'Koridor 3'), ('Jembatan Baru', 'Koridor 3'),
    ('Rawa Buaya', 'Koridor 3'), ('Sumur Bor', 'Koridor 3'), ('Pesakih', 'Koridor 3'),
    ('Kalideres', 'Koridor 3'),
    ('Pulo Gadung', 'Koridor 4'), ('Pasar Pulo Gadung', 'Koridor 4'),
    ('Pemuda Merdeka', 'Koridor 4'), ('Layur', 'Koridor 4'),
    ('Pemuda Rawamangun', 'Koridor 4'), ('Velodrome', 'Koridor 4'),
    ('Kayu Jati', 'Koridor 4'), ('Rawamangun', 'Koridor 4'),
    ('Pramuka Sari', 'Koridor 4'), ('Utan Kayu', 'Koridor 4'),
    ('Utan Kayu Rawamangun', 'Koridor 4'), ('Flyover Pramuka', 'Koridor 4'),
    ('Manggarai', 'Koridor 4'), ('Plaza Stasiun Manggarai', 'Koridor 4'),
    ('Pasar Rumput', 'Koridor 4'), ('Halimun', 'Koridor 4'), ('Galunggung', 'Koridor 4'),
    ('Ancol', 'Koridor 5'), ('Pademangan', 'Koridor 5'), ('Gunung Sahari', 'Koridor 5'),
    ('Jembatan Merah', 'Koridor 5'), ('Lapangan Banteng', 'Koridor 5'),
    ('Jaga Jakarta', 'Koridor 5'), ('Pal Putih', 'Koridor 5'),
    ('Kramat Sentiong', 'Koridor 5'), ('Salemba', 'Koridor 5'), ('Paseban', 'Koridor 5'),
    ('Matraman I', 'Koridor 5'), ('Tegalan', 'Koridor 5'), ('Matraman Baru', 'Koridor 5'),
    ('Bali Mester', 'Koridor 5'), ('Jatinegara', 'Koridor 5'), ('Kampung Melayu', 'Koridor 5'),
    ('Ragunan', 'Koridor 6'), ('Jati Barat', 'Koridor 6'), ('Jati Padang', 'Koridor 6'),
    ('Pejaten', 'Koridor 6'), ('Buncit Indah', 'Koridor 6'), ('Warung Jati', 'Koridor 6'),
    ('Warung Buncit', 'Koridor 6'), ('Duren Tiga', 'Koridor 6'),
    ('Mampang Prapatan', 'Koridor 6'), ('Underpass Kuningan', 'Koridor 6'),
    ('Patra Kuningan', 'Koridor 6'), ('Kuningan', 'Koridor 6'),
    ('Rasuna Said', 'Koridor 6'), ('Karet Kuningan', 'Koridor 6'),
    ('Kuningan Madya', 'Koridor 6'), ('Setiabudi', 'Koridor 6'),
    ('Halimun', 'Koridor 6'), ('Galunggung', 'Koridor 6'),
    ('Kampung Melayu', 'Koridor 7'), ('Bidara Cina', 'Koridor 7'),
    ('Gelanggang Remaja', 'Koridor 7'), ('Cawang Baru', 'Koridor 7'),
    ('BNN', 'Koridor 7'), ('Cawang Sentral', 'Koridor 7'),
    ('Cawang Cililitan', 'Koridor 7'), ('Cililitan', 'Koridor 7'),
    ('Trikora', 'Koridor 7'), ('Flyover Raya Bogor', 'Koridor 7'),
    ('Tanah Merdeka', 'Koridor 7'), ('Tanah Merdeka 2', 'Koridor 7'),
    ('Kampung Rambutan', 'Koridor 7'),
    ('Pasar Baru', 'Koridor 8'), ('Juanda', 'Koridor 8'), ('Pecenongan', 'Koridor 8'),
    ('Petojo', 'Koridor 8'), ('Tarakan', 'Koridor 8'), ('Tomang Raya', 'Koridor 8'),
    ('Tanjung Duren Arah P.Ranti', 'Koridor 8'), ('Tanjung Duren Arah Pluit', 'Koridor 8'),
    ('Grogol Reformasi', 'Koridor 8'), ('Jelambar', 'Koridor 8'), ('Damai', 'Koridor 8'),
    ('Kedoya', 'Koridor 8'), ('Kedoya Panjang', 'Koridor 8'), ('Duri Kepa', 'Koridor 8'),
    ('Kebon Jeruk', 'Koridor 8'), ('Kelapa Dua Sasak', 'Koridor 8'),
    ('Pos Pengumben', 'Koridor 8'), ('Arteri', 'Koridor 8'), ('Permata Hijau', 'Koridor 8'),
    ('Simprug', 'Koridor 8'), ('Kebayoran', 'Koridor 8'), ('Bungur', 'Koridor 8'),
    ('Tanah Kusir', 'Koridor 8'), ('Pondok Indah', 'Koridor 8'),
    ('Underpass Lebak Bulus', 'Koridor 8'), ('Pondok Pinang', 'Koridor 8'),
    ('Lebak Bulus', 'Koridor 8'),
    ('Jembatan Tiga', 'Koridor 9'), ('Jembatan Dua', 'Koridor 9'),
    ('Jembatan Besi', 'Koridor 9'), ('Kali Grogol Arah P.Ranti', 'Koridor 9'),
    ('Kali Grogol Arah Pluit', 'Koridor 9'), ('Grogol Reformasi', 'Koridor 9'),
    ('Tanjung Duren Arah P.Ranti', 'Koridor 9'), ('Tanjung Duren Arah Pluit', 'Koridor 9'),
    ('Kota Bambu Arah Pluit', 'Koridor 9'), ('Kota Bambu Arah P.Ranti', 'Koridor 9'),
    ('Kemanggisan Arah P.Ranti', 'Koridor 9'), ('Kemanggisan Arah Pluit', 'Koridor 9'),
    ('Petamburan', 'Koridor 9'), ('Gerbang Pemuda Arah P.Ranti', 'Koridor 9'),
    ('Gerbang Pemuda Arah Pluit', 'Koridor 9'), ('Semanggi', 'Koridor 9'),
    ('Widya Chandra Telkomsel Arah P.Ranti', 'Koridor 9'),
    ('Widya Chandra Telkomsel Arah Pluit', 'Koridor 9'),
    ('Denpasar Arah P.Ranti (Jamsostek)', 'Koridor 9'),
    ('Denpasar Arah Pluit (Jamsostek)', 'Koridor 9'),
    ('Tegal Parang Arah P.Ranti', 'Koridor 9'), ('Tegal Parang Arah Pluit', 'Koridor 9'),
    ('Pancoran Arah P.Ranti', 'Koridor 9'), ('Pancoran Arah Pluit', 'Koridor 9'),
    ('Pancoran Tugu', 'Koridor 9'), ('Tebet Eco Park Arah P.Ranti', 'Koridor 9'),
    ('Tebet Eco Park Arah Pluit', 'Koridor 9'), ('Cikoko Arah P.Ranti', 'Koridor 9'),
    ('Cikoko Arah Pluit', 'Koridor 9'), ('Ciliwung Arah P.Ranti', 'Koridor 9'),
    ('Ciliwung Arah Pluit', 'Koridor 9'), ('BNN', 'Koridor 9'), ('Cawang Sentral', 'Koridor 9'),
    ('Makasar', 'Koridor 9'), ('Pinang Ranti', 'Koridor 9'),
    ('Koja', 'Koridor 10'), ('Walikota Jakarta Utara Arah Cililitan', 'Koridor 10'),
    ('Walikota Jakarta Utara Arah Tj. Priuk', 'Koridor 10'), ('Plumpang', 'Koridor 10'),
    ('Sunter Kelapa Gading', 'Koridor 10'), ('Kodamar', 'Koridor 10'),
    ('Cempaka Putih', 'Koridor 10'), ('Pulo Mas Bypass', 'Koridor 10'),
    ('Kayu Putih Rawasari', 'Koridor 10'), ('Utan Kayu Rawamangun', 'Koridor 10'),
    ('Flyover Jatinegara', 'Koridor 10'), ('Pedati Prumpung', 'Koridor 10'),
    ('Kebon Nanas', 'Koridor 10'), ('Halim', 'Koridor 10'), ('Simpang Cawang', 'Koridor 10'),
    ('Cawang Sentral', 'Koridor 10'), ('Cawang Cililitan', 'Koridor 10'), ('PGC', 'Koridor 10'),
    ('Pulo Gebang', 'Koridor 11'), ('Walikota Jakarta Timur', 'Koridor 11'),
    ('Penggilingan', 'Koridor 11'), ('Flyover Pondok Kopi', 'Koridor 11'),
    ('Buaran', 'Koridor 11'), ('Kampung Sumur', 'Koridor 11'), ('Klender', 'Koridor 11'),
    ('Stasiun Klender', 'Koridor 11'), ('Cipinang', 'Koridor 11'),
    ('Flyover Cipinang', 'Koridor 11'), ('Pasar Enjo', 'Koridor 11'),
    ('Flyover Jatinegara', 'Koridor 11'), ('Stasiun Jatinegara', 'Koridor 11'),
    ('Jatinegara', 'Koridor 11'), ('Kampung Melayu', 'Koridor 11'),
    ('Pluit', 'Koridor 12'), ('Pluit Selatan', 'Koridor 12'), ('Pakin', 'Koridor 12'),
    ('Gedong Panjang', 'Koridor 12'), ('Museum Sejarah Jakarta', 'Koridor 12'),
    ('Kota', 'Koridor 12'), ('Mangga Dua Raya', 'Koridor 12'), ('Mangga Dua', 'Koridor 12'),
    ('Gunung Sahari', 'Koridor 12'), ('Jembatan Merah', 'Koridor 12'),
    ('Landasan Pacu', 'Koridor 12'), ('Danau Agung', 'Koridor 12'),
    ('Danau Sunter', 'Koridor 12'), ('Sunter Utara', 'Koridor 12'),
    ('Sunter Karya', 'Koridor 12'), ('Sunter Boulevard Barat', 'Koridor 12'),
    ('Sunter Kelapa Gading', 'Koridor 12'), ('Plumpang', 'Koridor 12'),
    ('Walikota Jakarta Timur', 'Koridor 12'), ('Koja', 'Koridor 12'),
    ('Penjaringan', 'Koridor 12'), ('Kali Besar', 'Koridor 12'),
    ('Puri Beta 2', 'Koridor 13'), ('Petukangan Utara', 'Koridor 13'),
    ('JORR', 'Koridor 13'), ('Swadarma', 'Koridor 13'), ('Cipulir', 'Koridor 13'),
    ('Seskoal', 'Koridor 13'), ('Kebayoran Lama', 'Koridor 13'), ('Velbak', 'Koridor 13'),
    ('Mayestik', 'Koridor 13'), ('CSW', 'Koridor 13'), ('ASEAN', 'Koridor 13'),
    ('Kejaksaan Agung', 'Koridor 13'), ('Pasar Santa', 'Koridor 13'),
    ('Rawa Barat', 'Koridor 13'), ('Tegal Mampang', 'Koridor 13'),
    ('Jembatan Item', 'Koridor 14'), ('Danau Sunter', 'Koridor 14'), ('Danau Agung', 'Koridor 14'),
    ('Landasan Pacu', 'Koridor 14'), ('Kemayoran', 'Koridor 14'), ('Tanah Tinggi', 'Koridor 14'),
    ('Senen Raya', 'Koridor 14')
]

# 1. Baca data dari Excel
df = pd.read_excel('/kaggle/input/datasets/elizarahmadania/tapin-tj-40-hari/(1) 40 HARI TAP IN 10 MENIT.xlsx')

# Ekstrak nama halte
halte_names = df['Nama Halte'].values

# Konstanta
INTERVALS_PER_DAY = 102
NUM_DAYS = 40
NUM_WEEKS = 8
DAYS_PER_WEEK = 5

# 2. Hitung daily totals
daily_totals = []
for day in range(NUM_DAYS):
    start_col = day * INTERVALS_PER_DAY
    end_col = (day + 1) * INTERVALS_PER_DAY
    day_data = df.iloc[:, 2 + start_col:2 + end_col].sum(axis=1).values
    daily_totals.append(day_data)

daily_totals = np.array(daily_totals).T  # Shape: (225 halte, 40 days)

# 3. Group per minggu
weekly_totals = []
for week in range(NUM_WEEKS):
    start_day = week * DAYS_PER_WEEK
    end_day = (week + 1) * DAYS_PER_WEEK
    week_data = daily_totals[:, start_day:end_day].sum(axis=1)
    weekly_totals.append(week_data)

weekly_totals = np.array(weekly_totals).T  # Shape: (225 halte, 8 weeks)

# 4. Buat dictionary untuk mapping halte ke list koridor
halte_to_corridors = {}
for halte, corridor in data_mapping:
    halte_clean = str(halte).upper().strip()
    if halte_clean not in halte_to_corridors:
        halte_to_corridors[halte_clean] = []
    halte_to_corridors[halte_clean].append(corridor)

# Group data by corridor
corridor_data = {}
for corridor in set([c[1] for c in data_mapping]):
    corridor_data[corridor] = {'halte': [], 'indices': []}

for idx, halte in enumerate(halte_names):
    halte_clean = str(halte).upper().strip()
    if halte_clean in halte_to_corridors:
        for corridor in halte_to_corridors[halte_clean]:
            corridor_data[corridor]['halte'].append(halte)
            corridor_data[corridor]['indices'].append(idx)

# 5. Hitung statistik per koridor & Lacak nama Halte
results = []
for corridor in sorted(corridor_data.keys(), key=lambda x: int(x.replace('Koridor ', ''))):
    indices = corridor_data[corridor]['indices']
    halte_list = corridor_data[corridor]['halte'] # Daftar nama halte khusus koridor ini

    corridor_weekly_data = weekly_totals[indices]
    num_halte = len(indices)

    if num_halte == 0:
        continue

    # Total volume seluruh halte di koridor ini per minggu
    volumes_per_week = corridor_weekly_data.sum(axis=0)

    # Hitung keseluruhan statistik DARI individu halte (tidak ditotalin)
    overall_max = np.max(corridor_weekly_data)
    overall_min = np.min(corridor_weekly_data)
    overall_range = overall_max - overall_min
    overall_median = np.median(corridor_weekly_data)
    overall_std = np.std(corridor_weekly_data, ddof=1) if corridor_weekly_data.size > 1 else 0

    # Lacak letak kordinat dari nilai max dan min di matriks 2D
    max_idx = np.unravel_index(np.argmax(corridor_weekly_data), corridor_weekly_data.shape)
    min_idx = np.unravel_index(np.argmin(corridor_weekly_data), corridor_weekly_data.shape)

    # Ambil nama halte berdasarkan kordinat baris (index ke-0 dari unravel_index)
    halte_max_name = halte_list[max_idx[0]]
    halte_min_name = halte_list[min_idx[0]]

    results.append({
        'Koridor': corridor,
        'Jml Halte': num_halte,
        'Vol W1': int(volumes_per_week[0]),
        'Vol W2': int(volumes_per_week[1]),
        'Vol W3': int(volumes_per_week[2]),
        'Vol W4': int(volumes_per_week[3]),
        'Vol W5': int(volumes_per_week[4]),
        'Vol W6': int(volumes_per_week[5]),
        'Vol W7': int(volumes_per_week[6]),
        'Vol W8': int(volumes_per_week[7]),
        'Nilai Max': int(overall_max),
        'Halte Max': halte_max_name,
        'Nilai Min': int(overall_min),
        'Halte Min': halte_min_name,
        'Range': int(overall_range),
        'Median': int(overall_median),
        'Std Dev': round(overall_std, 2)
    })

results_df = pd.DataFrame(results)

# 6. Menampilkan Hasil Langsung
# Setting pandas agar seluruh kolom & baris tercetak tanpa terpotong "..."
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)

print("\n" + "="*80)
print("HASIL ANALISIS KORIDOR TRANSAKARTA (40 HARI)".center(80))
print("="*80 + "\n")

# Perintah display ini akan memunculkan dataframe layaknya tabel cantik di Jupyter/Kaggle
try:
    display(results_df)
except NameError:
    # Fallback kalau misal dijalankan di terminal biasa bukan notebook
    print(results_df.to_string(index=False))
```

```python
import pandas as pd
import numpy as np

# Data mapping halte ke koridor
data_mapping = [
    ('Blok M', 'Koridor 1'), ('CSW', 'Koridor 1'), ('ASEAN', 'Koridor 1'),
    ('Kejaksaan Agung', 'Koridor 1'), ('Masjid Agung', 'Koridor 1'),
    ('Bundaran Senayan', 'Koridor 1'), ('Senayan Bank Jakarta', 'Koridor 1'),
    ('Polda Metro Jaya', 'Koridor 1'), ('Bendungan Hilir', 'Koridor 1'),
    ('Karet', 'Koridor 1'), ('Dukuh Atas', 'Koridor 1'), ('Tosari', 'Koridor 1'),
    ('Bundaran HI Astra', 'Koridor 1'), ('MH Thamrin', 'Koridor 1'),
    ('Kebon Sirih Arah Blok M', 'Koridor 1'), ('Kebon Sirih Arah Kota', 'Koridor 1'),
    ('Monumen Nasional', 'Koridor 1'), ('Harmoni Arah Blok M', 'Koridor 1'),
    ('Sawah Besar Arah Blok M', 'Koridor 1'), ('Taman Sari', 'Koridor 1'),
    ('Glodok', 'Koridor 1'), ('Kota', 'Koridor 1'), ('Kali Besar', 'Koridor 1'),
    ('Museum Sejarah Jakarta', 'Koridor 1'),
    ('Pulo Gadung', 'Koridor 2'), ('Bermis', 'Koridor 2'), ('Pulo Mas', 'Koridor 2'),
    ('Perintis Kemerdekaan', 'Koridor 2'), ('Pedongkelan', 'Koridor 2'),
    ('Cempaka Mas', 'Koridor 2'), ('Sumur Batu', 'Koridor 2'),
    ('Cempaka Baru', 'Koridor 2'), ('Pasar Cempaka Putih', 'Koridor 2'),
    ('Rawa Selatan', 'Koridor 2'), ('Galur', 'Koridor 2'), ('Senen Raya', 'Koridor 2'),
    ('RSPAD', 'Koridor 2'), ('Pejambon', 'Koridor 2'), ('Gambir', 'Koridor 2'),
    ('Istiqlal', 'Koridor 2'), ('Juanda', 'Koridor 2'), ('Pecenongan', 'Koridor 2'),
    ('Monumen Nasional', 'Koridor 2'), ('Balai Kota', 'Koridor 2'), ('Gambir 2', 'Koridor 2'),
    ('Kwitang', 'Koridor 2'),
    ('Monumen Nasional', 'Koridor 3'), ('Petojo', 'Koridor 3'), ('Roxy', 'Koridor 3'),
    ('Grogol', 'Koridor 3'), ('Jelambar', 'Koridor 3'), ('Damai', 'Koridor 3'),
    ('Taman Kota', 'Koridor 3'), ('Jembatan Gantung', 'Koridor 3'),
    ('Pulo Nangka', 'Koridor 3'), ('Jembatan Baru', 'Koridor 3'),
    ('Rawa Buaya', 'Koridor 3'), ('Sumur Bor', 'Koridor 3'), ('Pesakih', 'Koridor 3'),
    ('Kalideres', 'Koridor 3'),
    ('Pulo Gadung', 'Koridor 4'), ('Pasar Pulo Gadung', 'Koridor 4'),
    ('Pemuda Merdeka', 'Koridor 4'), ('Layur', 'Koridor 4'),
    ('Pemuda Rawamangun', 'Koridor 4'), ('Velodrome', 'Koridor 4'),
    ('Kayu Jati', 'Koridor 4'), ('Rawamangun', 'Koridor 4'),
    ('Pramuka Sari', 'Koridor 4'), ('Utan Kayu', 'Koridor 4'),
    ('Utan Kayu Rawamangun', 'Koridor 4'), ('Flyover Pramuka', 'Koridor 4'),
    ('Manggarai', 'Koridor 4'), ('Plaza Stasiun Manggarai', 'Koridor 4'),
    ('Pasar Rumput', 'Koridor 4'), ('Halimun', 'Koridor 4'), ('Galunggung', 'Koridor 4'),
    ('Ancol', 'Koridor 5'), ('Pademangan', 'Koridor 5'), ('Gunung Sahari', 'Koridor 5'),
    ('Jembatan Merah', 'Koridor 5'), ('Lapangan Banteng', 'Koridor 5'),
    ('Jaga Jakarta', 'Koridor 5'), ('Pal Putih', 'Koridor 5'),
    ('Kramat Sentiong', 'Koridor 5'), ('Salemba', 'Koridor 5'), ('Paseban', 'Koridor 5'),
    ('Matraman I', 'Koridor 5'), ('Tegalan', 'Koridor 5'), ('Matraman Baru', 'Koridor 5'),
    ('Bali Mester', 'Koridor 5'), ('Jatinegara', 'Koridor 5'), ('Kampung Melayu', 'Koridor 5'),
    ('Ragunan', 'Koridor 6'), ('Jati Barat', 'Koridor 6'), ('Jati Padang', 'Koridor 6'),
    ('Pejaten', 'Koridor 6'), ('Buncit Indah', 'Koridor 6'), ('Warung Jati', 'Koridor 6'),
    ('Warung Buncit', 'Koridor 6'), ('Duren Tiga', 'Koridor 6'),
    ('Mampang Prapatan', 'Koridor 6'), ('Underpass Kuningan', 'Koridor 6'),
    ('Patra Kuningan', 'Koridor 6'), ('Kuningan', 'Koridor 6'),
    ('Rasuna Said', 'Koridor 6'), ('Karet Kuningan', 'Koridor 6'),
    ('Kuningan Madya', 'Koridor 6'), ('Setiabudi', 'Koridor 6'),
    ('Halimun', 'Koridor 6'), ('Galunggung', 'Koridor 6'),
    ('Kampung Melayu', 'Koridor 7'), ('Bidara Cina', 'Koridor 7'),
    ('Gelanggang Remaja', 'Koridor 7'), ('Cawang Baru', 'Koridor 7'),
    ('BNN', 'Koridor 7'), ('Cawang Sentral', 'Koridor 7'),
    ('Cawang Cililitan', 'Koridor 7'), ('Cililitan', 'Koridor 7'),
    ('Trikora', 'Koridor 7'), ('Flyover Raya Bogor', 'Koridor 7'),
    ('Tanah Merdeka', 'Koridor 7'), ('Tanah Merdeka 2', 'Koridor 7'),
    ('Kampung Rambutan', 'Koridor 7'),
    ('Pasar Baru', 'Koridor 8'), ('Juanda', 'Koridor 8'), ('Pecenongan', 'Koridor 8'),
    ('Petojo', 'Koridor 8'), ('Tarakan', 'Koridor 8'), ('Tomang Raya', 'Koridor 8'),
    ('Tanjung Duren Arah P.Ranti', 'Koridor 8'), ('Tanjung Duren Arah Pluit', 'Koridor 8'),
    ('Grogol Reformasi', 'Koridor 8'), ('Jelambar', 'Koridor 8'), ('Damai', 'Koridor 8'),
    ('Kedoya', 'Koridor 8'), ('Kedoya Panjang', 'Koridor 8'), ('Duri Kepa', 'Koridor 8'),
    ('Kebon Jeruk', 'Koridor 8'), ('Kelapa Dua Sasak', 'Koridor 8'),
    ('Pos Pengumben', 'Koridor 8'), ('Arteri', 'Koridor 8'), ('Permata Hijau', 'Koridor 8'),
    ('Simprug', 'Koridor 8'), ('Kebayoran', 'Koridor 8'), ('Bungur', 'Koridor 8'),
    ('Tanah Kusir', 'Koridor 8'), ('Pondok Indah', 'Koridor 8'),
    ('Underpass Lebak Bulus', 'Koridor 8'), ('Pondok Pinang', 'Koridor 8'),
    ('Lebak Bulus', 'Koridor 8'),
    ('Jembatan Tiga', 'Koridor 9'), ('Jembatan Dua', 'Koridor 9'),
    ('Jembatan Besi', 'Koridor 9'), ('Kali Grogol Arah P.Ranti', 'Koridor 9'),
    ('Kali Grogol Arah Pluit', 'Koridor 9'), ('Grogol Reformasi', 'Koridor 9'),
    ('Tanjung Duren Arah P.Ranti', 'Koridor 9'), ('Tanjung Duren Arah Pluit', 'Koridor 9'),
    ('Kota Bambu Arah Pluit', 'Koridor 9'), ('Kota Bambu Arah P.Ranti', 'Koridor 9'),
    ('Kemanggisan Arah P.Ranti', 'Koridor 9'), ('Kemanggisan Arah Pluit', 'Koridor 9'),
    ('Petamburan', 'Koridor 9'), ('Gerbang Pemuda Arah P.Ranti', 'Koridor 9'),
    ('Gerbang Pemuda Arah Pluit', 'Koridor 9'), ('Semanggi', 'Koridor 9'),
    ('Widya Chandra Telkomsel Arah P.Ranti', 'Koridor 9'),
    ('Widya Chandra Telkomsel Arah Pluit', 'Koridor 9'),
    ('Denpasar Arah P.Ranti (Jamsostek)', 'Koridor 9'),
    ('Denpasar Arah Pluit (Jamsostek)', 'Koridor 9'),
    ('Tegal Parang Arah P.Ranti', 'Koridor 9'), ('Tegal Parang Arah Pluit', 'Koridor 9'),
    ('Pancoran Arah P.Ranti', 'Koridor 9'), ('Pancoran Arah Pluit', 'Koridor 9'),
    ('Pancoran Tugu', 'Koridor 9'), ('Tebet Eco Park Arah P.Ranti', 'Koridor 9'),
    ('Tebet Eco Park Arah Pluit', 'Koridor 9'), ('Cikoko Arah P.Ranti', 'Koridor 9'),
    ('Cikoko Arah Pluit', 'Koridor 9'), ('Ciliwung Arah P.Ranti', 'Koridor 9'),
    ('Ciliwung Arah Pluit', 'Koridor 9'), ('BNN', 'Koridor 9'), ('Cawang Sentral', 'Koridor 9'),
    ('Makasar', 'Koridor 9'), ('Pinang Ranti', 'Koridor 9'),
    ('Koja', 'Koridor 10'), ('Walikota Jakarta Utara Arah Cililitan', 'Koridor 10'),
    ('Walikota Jakarta Utara Arah Tj. Priuk', 'Koridor 10'), ('Plumpang', 'Koridor 10'),
    ('Sunter Kelapa Gading', 'Koridor 10'), ('Kodamar', 'Koridor 10'),
    ('Cempaka Putih', 'Koridor 10'), ('Pulo Mas Bypass', 'Koridor 10'),
    ('Kayu Putih Rawasari', 'Koridor 10'), ('Utan Kayu Rawamangun', 'Koridor 10'),
    ('Flyover Jatinegara', 'Koridor 10'), ('Pedati Prumpung', 'Koridor 10'),
    ('Kebon Nanas', 'Koridor 10'), ('Halim', 'Koridor 10'), ('Simpang Cawang', 'Koridor 10'),
    ('Cawang Sentral', 'Koridor 10'), ('Cawang Cililitan', 'Koridor 10'), ('PGC', 'Koridor 10'),
    ('Pulo Gebang', 'Koridor 11'), ('Walikota Jakarta Timur', 'Koridor 11'),
    ('Penggilingan', 'Koridor 11'), ('Flyover Pondok Kopi', 'Koridor 11'),
    ('Buaran', 'Koridor 11'), ('Kampung Sumur', 'Koridor 11'), ('Klender', 'Koridor 11'),
    ('Stasiun Klender', 'Koridor 11'), ('Cipinang', 'Koridor 11'),
    ('Flyover Cipinang', 'Koridor 11'), ('Pasar Enjo', 'Koridor 11'),
    ('Flyover Jatinegara', 'Koridor 11'), ('Stasiun Jatinegara', 'Koridor 11'),
    ('Jatinegara', 'Koridor 11'), ('Kampung Melayu', 'Koridor 11'),
    ('Pluit', 'Koridor 12'), ('Pluit Selatan', 'Koridor 12'), ('Pakin', 'Koridor 12'),
    ('Gedong Panjang', 'Koridor 12'), ('Museum Sejarah Jakarta', 'Koridor 12'),
    ('Kota', 'Koridor 12'), ('Mangga Dua Raya', 'Koridor 12'), ('Mangga Dua', 'Koridor 12'),
    ('Gunung Sahari', 'Koridor 12'), ('Jembatan Merah', 'Koridor 12'),
    ('Landasan Pacu', 'Koridor 12'), ('Danau Agung', 'Koridor 12'),
    ('Danau Sunter', 'Koridor 12'), ('Sunter Utara', 'Koridor 12'),
    ('Sunter Karya', 'Koridor 12'), ('Sunter Boulevard Barat', 'Koridor 12'),
    ('Sunter Kelapa Gading', 'Koridor 12'), ('Plumpang', 'Koridor 12'),
    ('Walikota Jakarta Timur', 'Koridor 12'), ('Koja', 'Koridor 12'),
    ('Penjaringan', 'Koridor 12'), ('Kali Besar', 'Koridor 12'),
    ('Puri Beta 2', 'Koridor 13'), ('Petukangan Utara', 'Koridor 13'),
    ('JORR', 'Koridor 13'), ('Swadarma', 'Koridor 13'), ('Cipulir', 'Koridor 13'),
    ('Seskoal', 'Koridor 13'), ('Kebayoran Lama', 'Koridor 13'), ('Velbak', 'Koridor 13'),
    ('Mayestik', 'Koridor 13'), ('CSW', 'Koridor 13'), ('ASEAN', 'Koridor 13'),
    ('Kejaksaan Agung', 'Koridor 13'), ('Pasar Santa', 'Koridor 13'),
    ('Rawa Barat', 'Koridor 13'), ('Tegal Mampang', 'Koridor 13'),
    ('Jembatan Item', 'Koridor 14'), ('Danau Sunter', 'Koridor 14'), ('Danau Agung', 'Koridor 14'),
    ('Landasan Pacu', 'Koridor 14'), ('Kemayoran', 'Koridor 14'), ('Tanah Tinggi', 'Koridor 14'),
    ('Senen Raya', 'Koridor 14')
]

# 1. Baca data dari Excel
df = pd.read_excel('/kaggle/input/datasets/elizarahmadania/tapout-tj-40-hari/(1) 40 HARI TAPOUT 10 MENIT.xlsx')

# Ekstrak nama halte
halte_names = df['Nama Halte'].values

# Konstanta
INTERVALS_PER_DAY = 102
NUM_DAYS = 40
NUM_WEEKS = 8
DAYS_PER_WEEK = 5

# 2. Hitung daily totals
daily_totals = []
for day in range(NUM_DAYS):
    start_col = day * INTERVALS_PER_DAY
    end_col = (day + 1) * INTERVALS_PER_DAY
    day_data = df.iloc[:, 2 + start_col:2 + end_col].sum(axis=1).values
    daily_totals.append(day_data)

daily_totals = np.array(daily_totals).T  # Shape: (225 halte, 40 days)

# 3. Group per minggu
weekly_totals = []
for week in range(NUM_WEEKS):
    start_day = week * DAYS_PER_WEEK
    end_day = (week + 1) * DAYS_PER_WEEK
    week_data = daily_totals[:, start_day:end_day].sum(axis=1)
    weekly_totals.append(week_data)

weekly_totals = np.array(weekly_totals).T  # Shape: (225 halte, 8 weeks)

# 4. Buat dictionary untuk mapping halte ke list koridor
halte_to_corridors = {}
for halte, corridor in data_mapping:
    halte_clean = str(halte).upper().strip()
    if halte_clean not in halte_to_corridors:
        halte_to_corridors[halte_clean] = []
    halte_to_corridors[halte_clean].append(corridor)

# Group data by corridor
corridor_data = {}
for corridor in set([c[1] for c in data_mapping]):
    corridor_data[corridor] = {'halte': [], 'indices': []}

for idx, halte in enumerate(halte_names):
    halte_clean = str(halte).upper().strip()
    if halte_clean in halte_to_corridors:
        for corridor in halte_to_corridors[halte_clean]:
            corridor_data[corridor]['halte'].append(halte)
            corridor_data[corridor]['indices'].append(idx)

# 5. Hitung statistik per koridor & Lacak nama Halte
results = []
for corridor in sorted(corridor_data.keys(), key=lambda x: int(x.replace('Koridor ', ''))):
    indices = corridor_data[corridor]['indices']
    halte_list = corridor_data[corridor]['halte'] # Daftar nama halte khusus koridor ini

    corridor_weekly_data = weekly_totals[indices]
    num_halte = len(indices)

    if num_halte == 0:
        continue

    # Total volume seluruh halte di koridor ini per minggu
    volumes_per_week = corridor_weekly_data.sum(axis=0)

    # Hitung keseluruhan statistik DARI individu halte (tidak ditotalin)
    overall_max = np.max(corridor_weekly_data)
    overall_min = np.min(corridor_weekly_data)
    overall_range = overall_max - overall_min
    overall_median = np.median(corridor_weekly_data)
    overall_std = np.std(corridor_weekly_data, ddof=1) if corridor_weekly_data.size > 1 else 0

    # Lacak letak kordinat dari nilai max dan min di matriks 2D
    max_idx = np.unravel_index(np.argmax(corridor_weekly_data), corridor_weekly_data.shape)
    min_idx = np.unravel_index(np.argmin(corridor_weekly_data), corridor_weekly_data.shape)

    # Ambil nama halte berdasarkan kordinat baris (index ke-0 dari unravel_index)
    halte_max_name = halte_list[max_idx[0]]
    halte_min_name = halte_list[min_idx[0]]

    results.append({
        'Koridor': corridor,
        'Jml Halte': num_halte,
        'Vol W1': int(volumes_per_week[0]),
        'Vol W2': int(volumes_per_week[1]),
        'Vol W3': int(volumes_per_week[2]),
        'Vol W4': int(volumes_per_week[3]),
        'Vol W5': int(volumes_per_week[4]),
        'Vol W6': int(volumes_per_week[5]),
        'Vol W7': int(volumes_per_week[6]),
        'Vol W8': int(volumes_per_week[7]),
        'Nilai Max': int(overall_max),
        'Halte Max': halte_max_name,
        'Nilai Min': int(overall_min),
        'Halte Min': halte_min_name,
        'Range': int(overall_range),
        'Median': int(overall_median),
        'Std Dev': round(overall_std, 2)
    })

results_df = pd.DataFrame(results)

# 6. Menampilkan Hasil Langsung
# Setting pandas agar seluruh kolom & baris tercetak tanpa terpotong "..."
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)

print("\n" + "="*80)
print("HASIL ANALISIS KORIDOR TRANSAKARTA (40 HARI) TAPOUT".center(80))
print("="*80 + "\n")

# Perintah display ini akan memunculkan dataframe layaknya tabel cantik di Jupyter/Kaggle
try:
    display(results_df)
except NameError:
    # Fallback kalau misal dijalankan di terminal biasa bukan notebook
    print(results_df.to_string(index=False))
```

```python
# KONFIGURASI PARAMETER (Sesuai Batasan)

STATION_NUMBER = 225
TOTAL_DAYS = 40    # 8 Minggu Hari Kerja (Senin-Jumat)
TRAIN_DAYS = 30    # 6 Minggu Hari Kerja
TEST_DAYS = 10     # 2 Minggu Hari Kerja

INTERVAL_CONFIG = {
    10: {
        'tg_per_day': 102,
        'path_in': '/kaggle/input/datasets/elizarahmadania/tapin-tj-40-hari/(1) 40 HARI TAP IN 10 MENIT.xlsx',
        'path_out': '/kaggle/input/datasets/elizarahmadania/tapout-tj-40-hari/(1) 40 HARI TAPOUT 10 MENIT.xlsx'
    },
    15: {
        'tg_per_day': 68,
        'path_in': '/kaggle/input/datasets/elizarahmadania/tapin-tj-40-hari/(1) 40 HARI TAP IN 15 MENIT.xlsx',
        'path_out': '/kaggle/input/datasets/elizarahmadania/tapout-tj-40-hari/(1) 40 HARI TAPOUT 15 MENIT.xlsx'
    },
    30: {
        'tg_per_day': 34,
        'path_in': '/kaggle/input/datasets/elizarahmadania/tapin-tj-40-hari/(1) 40 HARI TAP IN 30 MENIT.xlsx',
        'path_out': '/kaggle/input/datasets/elizarahmadania/tapout-tj-40-hari/(1) 40 HARI TAPOUT 30 MENIT.xlsx'
    }
}

HYPERPARAMS = {
    'time_lag': [10, 20, 30],
    'batch_size': [4, 8, 16, 32, 64, 128, 256],
    'conv_filters': [1, 2, 4, 8, 16, 32, 64],
    'gcn_units': [6, 9, 12, 15, 18, 21, 24]
}
```

### Utils

```python
# FUNGSI UTILITAS GRAPH & GCN LAYER

def normalized_adj(adj):
    adj = sp.coo_matrix(adj)
    rowsum = np.array(adj.sum(1))
    d_inv_sqrt = np.power(rowsum, -0.5).flatten()
    d_inv_sqrt[np.isinf(d_inv_sqrt)] = 0.
    d_mat_inv_sqrt = sp.diags(d_inv_sqrt)
    return adj.dot(d_mat_inv_sqrt).transpose().dot(d_mat_inv_sqrt).tocoo().astype(np.float32)

def sparse_to_tuple(mx):
    if not sp.isspmatrix_coo(mx): mx = mx.tocoo()
    coords = np.vstack((mx.row, mx.col)).transpose()
    return tf.sparse.SparseTensor(coords, mx.data, mx.shape)

def calculate_laplacian(adj):
    adj_normalized = normalized_adj(adj)
    laplacian = sp.eye(adj_normalized.shape[0]) - adj_normalized
    return sparse_to_tuple(laplacian)

class GraphConvolution1(Layer):
    def __init__(self, units, adj=None, activation=None, **kwargs):
        super(GraphConvolution1, self).__init__(**kwargs)
        self.adj = tf.sparse.to_dense(calculate_laplacian(adj))
        self.units = units
        self.activation = activations.get(activation)

    def build(self, input_shapes):
        self.kernel = self.add_weight(shape=(input_shapes[2], self.units), initializer='glorot_uniform', name='kernel')
        self.bias = self.add_weight(shape=(self.units,), initializer='zeros', name='bias')

    def call(self, inputs):
        supports = K.dot(tf.cast(self.adj, dtype=inputs.dtype), inputs)
        supports = tf.transpose(supports, perm=[1, 0, 2])
        output = K.dot(supports, self.kernel) + self.bias
        return self.activation(output)
```

### Load Data dan Split Data

```python
# FUNGSI LOADING, CLEANING, & SPLIT

def create_windows(data_in, data_out, time_lag):
    X1, X2, Y = [], [], []
    for i in range(len(data_in) - time_lag):
        X1.append(data_in[i : i + time_lag, :].T) # Shape: (Nodes, TimeLag)
        X2.append(data_out[i : i + time_lag, :].T)
        Y.append(data_in[i + time_lag, :])        # Target Inflow t+1
    return np.array(X1), np.array(X2), np.array(Y)

def load_and_clean_excel(path, tg_per_day):
    # Membaca Excel
    df = pd.read_excel(path)

    # Ambil data numerik (Baris 0-224, Kolom 2 sampai akhir)
    # Kolom 0 & 1 adalah Kode & Nama Halte (Metadata)
    data_numeric = df.iloc[:STATION_NUMBER, 2:].values

    # Transpose agar Baris = Waktu, Kolom = Halte
    data_numeric = data_numeric.T

    # Potong agar pas 40 hari (mencegah kelebihan baris kosong/tambahan di excel)
    expected_rows = TOTAL_DAYS * tg_per_day
    data_numeric = data_numeric[:expected_rows, :].astype(float)

    return data_numeric

def Get_All_Data(TG, time_lag, TG_in_one_day):
    path_in = INTERVAL_CONFIG[TG]['path_in']
    path_out = INTERVAL_CONFIG[TG]['path_out']

    # Load & Transpose
    metro_enter = load_and_clean_excel(path_in, TG_in_one_day)
    metro_exit = load_and_clean_excel(path_out, TG_in_one_day)

    # Normalisasi Min-Max Global
    a, b = np.max(metro_enter), np.min(metro_enter)
    metro_enter_norm = (metro_enter - b) / (a - b)
    metro_exit_norm = (metro_exit - b) / (a - b)

    # Splitting Point (6 Minggu = 30 Hari Kerja)
    split_point = TRAIN_DAYS * TG_in_one_day

    # Data Latih & Uji
    train_in_raw = metro_enter_norm[:split_point, :]
    train_out_raw = metro_exit_norm[:split_point, :]

    # Ambil sedikit data dari akhir training agar window testing pertama lengkap
    test_in_raw = metro_enter_norm[split_point - time_lag:, :]
    test_out_raw = metro_exit_norm[split_point - time_lag:, :]

    X_train_1, X_train_2, Y_train = create_windows(train_in_raw, train_out_raw, time_lag)
    X_test_1, X_test_2, Y_test = create_windows(test_in_raw, test_out_raw, time_lag)

    Y_test_original = Y_test * (a - b) + b

    return X_train_1, Y_train, X_test_1, Y_test, Y_test_original, a, b, X_train_2, X_test_2
```

### Evaluasi dan Pembentukan Model

```python
# EVALUASI & MODEL BUILDING

def evaluate_performance(Y_true, Y_pred):
    RMSE = sqrt(mean_squared_error(Y_true, Y_pred))
    R2 = r2_score(Y_true, Y_pred)
    MAE = mean_absolute_error(Y_true, Y_pred)
    # WMAPE
    total_sum = np.sum(Y_true)
    WMAPE = np.sum(np.abs(Y_true - Y_pred)) / total_sum if total_sum > 0 else 0
    return RMSE, R2, MAE, WMAPE

def build_dynamic_model(time_lag, gcn_units, conv_filters, adj_matrix):
    input1 = Input(shape=(STATION_NUMBER, time_lag))
    input2 = Input(shape=(STATION_NUMBER, time_lag))

    out1 = GraphConvolution1(gcn_units, adj=adj_matrix, activation='relu')(input1)
    out1 = Reshape((STATION_NUMBER, gcn_units, 1))(out1)

    out2 = GraphConvolution1(gcn_units, adj=adj_matrix, activation='relu')(input2)
    out2 = Reshape((STATION_NUMBER, gcn_units, 1))(out2)

    merged = concatenate([out1, out2], axis=-1)
    out = Conv2D(conv_filters, kernel_size=(3, 3), padding='same', activation='relu')(merged)
    out = Flatten()(out)
    out = Dense(STATION_NUMBER)(out)

    model = Model(inputs=[input1, input2], outputs=out)
    model.compile(loss='mse', optimizer=Adam(learning_rate=0.001))
    return model
```

```python
#  EKSEKUSI UTAMA

adj_path = "/kaggle/input/datasets/elizarahmadania/matrix-ketetanggaan-tj/matrix adjacency.xlsx"
df_adj = pd.read_excel(adj_path)

adjacency_data = df_adj.iloc[1:1+STATION_NUMBER, 2:2+STATION_NUMBER].values

# Konversi ke float sekarang akan berhasil karena hanya berisi angka
adjacency = adjacency_data.astype(float) + np.eye(STATION_NUMBER)

print(f"Matriks Adjacency Berhasil Dimuat: {adjacency.shape}") # Harus (225, 225)
```

### Interval Waktu Terbaik

```python
# MENCARI INTERVAL WAKTU TERBAIK (DEFAULT PARAMS)

default_params = {'time_lag': 10, 'batch_size': 32, 'conv_filters': 16, 'gcn_units': 18}
interval_scores = {}

print("=== TAHAP 1: SELEKSI INTERVAL TERBAIK ===")
for interval, config in INTERVAL_CONFIG.items():
    print(f"Mengevaluasi Interval {interval} Menit...")
    X_tr1, Y_tr, X_te1, Y_te, Y_te_orig, a, b, X_tr2, X_te2 = Get_All_Data(
        interval, default_params['time_lag'], config['tg_per_day']
    )

    tf.keras.backend.clear_session()
    model = build_dynamic_model(default_params['time_lag'], default_params['gcn_units'], default_params['conv_filters'], adjacency)
    model.fit([X_tr1, X_tr2], Y_tr, batch_size=default_params['batch_size'], epochs=15, verbose=0)

    y_pred = model.predict([X_te1, X_te2])
    y_pred_denorm = y_pred * (a - b) + b
    rmse, r2, mae, wmape = evaluate_performance(Y_te_orig, y_pred_denorm)

    interval_scores[interval] = wmape # Simpan WMAPE untuk perbandingan
    print(f"   > Hasil: WMAPE = {wmape:.4f}, R2 = {r2:.4f}")

# Cari interval dengan WMAPE terkecil
best_interval = min(interval_scores, key=interval_scores.get)
print(f"\n INTERVAL TERPILIH: {best_interval} Menit (WMAPE terendah)")
```

### Hyperparameter Tuning

```python
# TUNING HYPERPARAMETER

best_params = default_params.copy()
config_best = INTERVAL_CONFIG[best_interval]

print(f"\n=== TUNING PADA INTERVAL {best_interval} MENIT ===")

# Tune satu per satu, kunci yang terbaik
for param in ['time_lag', 'batch_size', 'conv_filters', 'gcn_units']:
    print(f"\nTuning {param}...")
    best_rmse_val = float('inf')
    best_value_found = best_params[param]

    for val in HYPERPARAMS[param]:
        test_params = best_params.copy()
        test_params[param] = val

        try:
            X_tr1, Y_tr, X_te1, Y_te, Y_te_orig, a, b, X_tr2, X_te2 = Get_All_Data(
                best_interval, test_params['time_lag'], config_best['tg_per_day']
            )

            tf.keras.backend.clear_session()
            model = build_dynamic_model(test_params['time_lag'], test_params['gcn_units'], test_params['conv_filters'], adjacency)
            model.fit([X_tr1, X_tr2], Y_tr, batch_size=test_params['batch_size'], epochs=10, verbose=0)

            y_pred = model.predict([X_te1, X_te2])
            y_pred_denorm = y_pred * (a - b) + b
            rmse_test, _, _, _ = evaluate_performance(Y_te_orig, y_pred_denorm)

            print(f"   Mencoba {val}: RMSE = {rmse_test:.4f}")
            if rmse_test < best_rmse_val:
                best_rmse_val = rmse_test
                best_value_found = val
        except Exception as e:
            print(f"   Error pada {val}: {e}")

    best_params[param] = best_value_found
    print(f"固定 (Fixed) {param} = {best_value_found}")
```

### Pelatihan Model

```python
#  PELATIHAN FINAL (MODEL OPTIMAL)

print(f"\n=== PELATIHAN FINAL (Interval {best_interval}) ===")
X_tr1, Y_tr, X_te1, Y_te, Y_te_orig, a, b, X_tr2, X_te2 = Get_All_Data(
    best_interval, best_params['time_lag'], config_best['tg_per_day']
)

tf.keras.backend.clear_session()
final_model = build_dynamic_model(best_params['time_lag'], best_params['gcn_units'], best_params['conv_filters'], adjacency)
history = final_model.fit([X_tr1, X_tr2], Y_tr, batch_size=best_params['batch_size'], epochs=100, verbose=1)

# Evaluasi Akhir
y_pred_final = final_model.predict([X_te1, X_te2])
y_pred_denorm_final = y_pred_final * (a - b) + b
rmse, r2, mae, wmape = evaluate_performance(Y_te_orig, y_pred_denorm_final)

print("\n" + "="*45)
print("HASIL AKHIR PENELITIAN")
print(f"Interval Terbaik      : {best_interval} Menit")
print(f"Hyperparams Optimal   : {best_params}")
print(f"RMSE                  : {rmse:.4f}")
print(f"R-Squared (R2)        : {r2:.4f}")
print(f"WMAPE                 : {wmape:.4f}")
print("="*45)
```

### Load, Cleaning, and Windowing

```python
# FUNGSI LOADING, CLEANING, & WINDOWING

def create_windows(data_in, data_out, time_lag):
    X1, X2, Y = [], [], []
    for i in range(len(data_in) - time_lag):
        X1.append(data_in[i : i + time_lag, :].T)
        X2.append(data_out[i : i + time_lag, :].T)
        Y.append(data_in[i + time_lag, :])
    return np.array(X1), np.array(X2), np.array(Y)

def load_and_clean_excel(path, tg_per_day):
    df = pd.read_excel(path)
    data_numeric = df.iloc[:STATION_NUMBER, 2:].values
    data_numeric = data_numeric.T
    expected_rows = TOTAL_DAYS * tg_per_day
    data_numeric = data_numeric[:expected_rows, :].astype(float)
    return data_numeric

def extract_data_by_days(enter_norm, exit_norm, start_day, end_day, tg_per_day, time_lag):
    # Disesuaikan dengan lag agar labelnya tepat mulai dari start_day
    start_idx = max(0, start_day * tg_per_day - time_lag)
    end_idx = end_day * tg_per_day

    d_in = enter_norm[start_idx:end_idx, :]
    d_out = exit_norm[start_idx:end_idx, :]
    return create_windows(d_in, d_out, time_lag)

def load_raw_data(interval, config):
    path_in = config['path_in']
    path_out = config['path_out']
    tg_per_day = config['tg_per_day']

    metro_enter = load_and_clean_excel(path_in, tg_per_day)
    metro_exit = load_and_clean_excel(path_out, tg_per_day)

    a, b = np.max(metro_enter), np.min(metro_enter)
    metro_enter_norm = (metro_enter - b) / (a - b)
    metro_exit_norm = (metro_exit - b) / (a - b)

    return metro_enter_norm, metro_exit_norm, a, b
```

## Hyperparameter Tuning Expanding Window

```python
# TUNING HYPERPARAMETER DENGAN EXPANDING WINDOW (4 FOLD)

config_best = INTERVAL_CONFIG[best_interval]
en_norm, ex_norm, a, b = load_raw_data(best_interval, config_best)
tg_per_day = config_best['tg_per_day']

# Custom Splits 4-Fold Validation
# Training Awal: 3 Minggu (Hari 0-15)
# Val 1: Minggu 4 (Hari 15-20), Val 2: M5 (20-25), Val 3: M6 (25-30), Val 4: M7 (30-35)
folds = [
    {'val_s': 15, 'val_e': 20},
    {'val_s': 20, 'val_e': 25},
    {'val_s': 25, 'val_e': 30},
    {'val_s': 30, 'val_e': 35}
]

best_params = default_params.copy()
print(f"\n=== TUNING PADA INTERVAL {best_interval} MENIT (EXPANDING WINDOW) ===")

for param in ['time_lag', 'batch_size', 'conv_filters', 'gcn_units']:
    print(f"\nTuning {param}...")
    best_rmse_val = float('inf')
    best_value_found = best_params[param]

    for val in HYPERPARAMS[param]:
        test_params = best_params.copy()
        test_params[param] = val
        lag = test_params['time_lag']

        fold_rmses = []
        for i, fold in enumerate(folds):
            # Terapkan Expanding Window (Latih dari awal 0 hingga awal titik fold)
            X_tr1, X_tr2, Y_tr = extract_data_by_days(en_norm, ex_norm, 0, fold['val_s'], tg_per_day, lag)
            X_val1, X_val2, Y_val = extract_data_by_days(en_norm, ex_norm, fold['val_s'], fold['val_e'], tg_per_day, lag)

            Y_val_orig = Y_val * (a - b) + b

            tf.keras.backend.clear_session()
            model = build_dynamic_model(lag, test_params['gcn_units'], test_params['conv_filters'], adjacency)
            model.fit([X_tr1, X_tr2], Y_tr, batch_size=test_params['batch_size'], epochs=5, verbose=0) # Epoch kecil utk CV

            y_pred = model.predict([X_val1, X_val2], verbose=0)
            y_pred_denorm = y_pred * (a - b) + b
            rmse_val, _, _, _ = evaluate_performance(Y_val_orig, y_pred_denorm)
            fold_rmses.append(rmse_val)

        avg_rmse = np.mean(fold_rmses)
        print(f"   Mencoba {val}: Avg Val RMSE = {avg_rmse:.4f}")

        if avg_rmse < best_rmse_val:
            best_rmse_val = avg_rmse
            best_value_found = val

    best_params[param] = best_value_found
    print(f"[*] Fixed {param} = {best_value_found}")
```

### Pelatihan dan Evaluasi

```python
# PELATIHAN FINAL & EVALUASI

print(f"\n=== PELATIHAN FINAL (Interval {best_interval} Menit) ===")
# Data Latih Semua Hari Kecuali 1 Minggu Terakhir (0 - 35)
X_tr1, X_tr2, Y_tr = extract_data_by_days(en_norm, ex_norm, 0, 35, tg_per_day, best_params['time_lag'])
# Data Test Hanya 1 Minggu Terakhir (35 - 40)
X_te1, X_te2, Y_te = extract_data_by_days(en_norm, ex_norm, 35, 40, tg_per_day, best_params['time_lag'])
Y_te_orig = Y_te * (a - b) + b

tf.keras.backend.clear_session()
final_model = build_dynamic_model(best_params['time_lag'], best_params['gcn_units'], best_params['conv_filters'], adjacency)
history = final_model.fit([X_tr1, X_tr2], Y_tr, batch_size=best_params['batch_size'], epochs=100, verbose=1)

y_pred_final = final_model.predict([X_te1, X_te2])
y_pred_denorm_final = y_pred_final * (a - b) + b
rmse, r2, mae, wmape = evaluate_performance(Y_te_orig, y_pred_denorm_final)

print("\n" + "="*45)
print("HASIL AKHIR PENELITIAN")
print(f"Interval Terbaik      : {best_interval} Menit")
print(f"Hyperparams Optimal   : {best_params}")
print(f"RMSE                  : {rmse:.4f}")
print(f"R-Squared (R2)        : {r2:.4f}")
print(f"WMAPE                 : {wmape:.4f}")
print("="*45)
```
